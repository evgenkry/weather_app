```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Реальные средние температуры (примерные данные) для городов по сезонам
seasonal_temperatures = {
    "New York": {"winter": 0, "spring": 10, "summer": 25, "autumn": 15},
    "London": {"winter": 5, "spring": 11, "summer": 18, "autumn": 12},
    "Paris": {"winter": 4, "spring": 12, "summer": 20, "autumn": 13},
    "Tokyo": {"winter": 6, "spring": 15, "summer": 27, "autumn": 18},
    "Moscow": {"winter": -10, "spring": 5, "summer": 18, "autumn": 8},
    "Sydney": {"winter": 12, "spring": 18, "summer": 25, "autumn": 20},
    "Berlin": {"winter": 0, "spring": 10, "summer": 20, "autumn": 11},
    "Beijing": {"winter": -2, "spring": 13, "summer": 27, "autumn": 16},
    "Rio de Janeiro": {"winter": 20, "spring": 25, "summer": 30, "autumn": 25},
    "Dubai": {"winter": 20, "spring": 30, "summer": 40, "autumn": 30},
    "Los Angeles": {"winter": 15, "spring": 18, "summer": 25, "autumn": 20},
    "Singapore": {"winter": 27, "spring": 28, "summer": 28, "autumn": 27},
    "Mumbai": {"winter": 25, "spring": 30, "summer": 35, "autumn": 30},
    "Cairo": {"winter": 15, "spring": 25, "summer": 35, "autumn": 25},
    "Mexico City": {"winter": 12, "spring": 18, "summer": 20, "autumn": 15},
}

# Сопоставление месяцев с сезонами
month_to_season = {12: "winter", 1: "winter", 2: "winter",
                   3: "spring", 4: "spring", 5: "spring",
                   6: "summer", 7: "summer", 8: "summer",
                   9: "autumn", 10: "autumn", 11: "autumn"}

# Генерация данных о температуре
def generate_realistic_temperature_data(cities, num_years=10):
    dates = pd.date_range(start="2010-01-01", periods=365 * num_years, freq="D")
    data = []

    for city in cities:
        for date in dates:
            season = month_to_season[date.month]
            mean_temp = seasonal_temperatures[city][season]
            # Добавляем случайное отклонение
            temperature = np.random.normal(loc=mean_temp, scale=5)
            data.append({"city": city, "timestamp": date, "temperature": temperature})

    df = pd.DataFrame(data)
    df['season'] = df['timestamp'].dt.month.map(lambda x: month_to_season[x])
    return df

# Генерация данных
data = generate_realistic_temperature_data(list(seasonal_temperatures.keys()))
data.to_csv('temperature_data.csv', index=False)
```

```python
data = data.set_index('timestamp')

def calculate_rolling_mean(df, window=30):
    rolling_means = {}
    for city in df['city'].unique():
        city_data = df[df['city'] == city].sort_index()
        rolling_mean = city_data['temperature'].rolling(window=window, center=False).mean()
        rolling_means[city] = rolling_mean
    return rolling_means

rolling_means = calculate_rolling_mean(data)

data['rolling_mean_30'] = np.nan

for city, rm in rolling_means.items():
    data.loc[data['city'] == city, 'rolling_mean_30'] = rm

data = data.reset_index()

data.head()
```

```python
plt.figure(figsize=(15, 8))

for city in data['city'].unique():
    city_data = data[data['city'] == city].sort_values('timestamp')
    plt.plot(city_data['timestamp'], city_data['rolling_mean_30'], label=city, alpha=0.9)

plt.xlabel('Дата', fontsize=12)
plt.ylabel('Температура (°C)', fontsize=12)
plt.title('Cкользящее среднее температуры по городам', fontsize=14)
plt.legend(loc='upper right', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.6)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
def calculate_seasonal_stats(df):
    seasonal_stats = df.groupby(['city', 'season'])['temperature'].agg(['mean', 'std']).reset_index()
    seasonal_stats = seasonal_stats.rename(columns={'mean': 'mean_temperature', 'std': 'std_temperature'})
    return seasonal_stats

seasonal_stats = calculate_seasonal_stats(data)
seasonal_stats.head()
```

```python
data = pd.merge(data, seasonal_stats, on=['city', 'season'], how='left')

data['upper_bound'] = data['mean_temperature'] + 2 * data['std_temperature']
data['lower_bound'] = data['mean_temperature'] - 2 * data['std_temperature']

data['is_anomaly'] = (data['temperature'] < data['lower_bound']) | (data['temperature'] > data['upper_bound'])

data.groupby('city')['is_anomaly'].sum().sort_values(ascending=False)
```

```python
from datetime import datetime

start_time = datetime.now()

cities = data['city'].unique().tolist()

for city in cities:
  city_data = data[data["city"] == city]
  curr_date = datetime.now()
  curr_season = month_to_season[curr_date.month]
  city_season = city_data[city_data["season"] == curr_season]

  start = datetime.now()
  curr_temperature = get_temperature(city, api_key)
  end = datetime.now()

  resp_time = (end - start).total_seconds()

  if curr_temperature is not None:
    lower = city_season["lower_bound"].iloc[0]
    upper = city_season["upper_bound"].iloc[0]
    if lower <= curr_temperature <= upper:
      print(f"{city}: {curr_temperature} °C, нормальная. Время ответа: {resp_time:.2f} сек.")
    else:
      print(f"{city}: {curr_temperature} °C, аномальная. Время ответа: {resp_time:.2f} сек.")
  else:
    print(f"Не удалось получить температуру для {city}")

end_time = datetime.now()
total_sync = (end_time - start_time).total_seconds()
print("-----------")
print(f"Общее время выполнения: {total_sync:.2f} сек.")
```

```python
import aiohttp
import asyncio
import nest_asyncio

async def get_temperature_async(city, api_key):
  url = f"https://api.openweathermap.org/data/2.5/weather?q={city}&appid={api_key}&units=metric"
  start = datetime.now()

  async with aiohttp.ClientSession() as session:
    async with session.get(url) as response:
      end = datetime.now()
      resp_time = (end - start).total_seconds()

      if response.status == 200:
        data = await response.json()
        temperature = data["main"]["temp"]
        return temperature, resp_time
      else:
        print(f"Ошибка: {response.status}, {await response.text()}")
        return None, resp_time

total_async = None

async def main(api_key, cities, data):
  global total_async

  overall_start_time = datetime.now()

  tasks = [get_temperature_async(city, api_key) for city in cities]

  temperatures = await asyncio.gather(*tasks)

  for city, (curr_temperature, resp_time) in zip(cities, temperatures):
    city_data = data[data["city"] == city]
    curr_date = datetime.now()
    curr_season = month_to_season[curr_date.month]
    city_season = city_data[city_data["season"] == curr_season]

    if curr_temperature is not None:
      lower = city_season["lower_bound"].iloc[0]
      upper = city_season["upper_bound"].iloc[0]
      if lower <= curr_temperature <= upper:
        print(f"{city}: {curr_temperature} °C, нормальная. Время ответа: {resp_time:.2f} сек.")
      else:
        print(f"{city}: {curr_temperature} °C, аномальная. Время ответа: {resp_time:.2f} сек.")
    else:
      print(f"Не удалось получить температуру для {city}. Время ответа: {resp_time:.2f} сек.")

  overall_end_time = datetime.now()
  total_async = (overall_end_time - overall_start_time).total_seconds()
  print("-----------")
  print(f"Общее время выполнения: {total_async:.2f} сек.")

nest_asyncio.apply()
await main(api_key, cities, data)
```

```python
diff =  total_sync / total_async
print(f"Асинхронно запросы выполняются в {diff:.2f} раз быстрее. Круто!")
```
